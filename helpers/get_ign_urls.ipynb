```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Get all links from the page
# https://www.ign.com/reviews/games/nintendo-switch
url = 'https://www.ign.com/reviews/games/nintendo-switch'
browser = webdriver.Chrome(ChromeDriverManager().install())
browser.get(url)
page = browser.find_element(By.XPATH, '/html')
```

```python
for _ in range(30):
    page.send_keys(Keys.SPACE)
    time.sleep(1.5)
```

```python
games = browser.find_elements(By.XPATH, '//a[@class="item-body"]')
games = [(game.get_attribute('aria-label'), game.get_attribute('href')) for game in games]
```

```python
for game in games:
    print(game[0], '|', game[1])
```
